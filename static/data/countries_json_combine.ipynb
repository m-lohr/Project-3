```python
import pandas as pd
import json
from urllib.request import urlopen
```

```python

# load json file using json.load
with open('countries.geojson') as file:
    polygon_data = json.load(file)
```

```python
# create dataframe
polygon_df = pd.json_normalize(polygon_data['features'])
polygon_df
```

```python
clean_polygon_df = polygon_df[['properties.ADMIN', 'geometry.type', 'geometry.coordinates']].copy()
clean_polygon_df.rename(columns={'properties.ADMIN':'country'}, inplace=True)
clean_polygon_df
```

```python
# load json file using json.load
with open('all_population.json') as file:
    population_data = json.load(file)

print(population_data)
```

```python
# create dataframe
population_df = pd.json_normalize(population_data)
population_df
```

```python
clean_population_df = population_df[['name.common', 'population']].copy()
clean_population_df.rename(columns={'name.common':'country'}, inplace=True)
clean_population_df
```

```python
clean_population_df = clean_population_df.merge(clean_polygon_df, how='inner', on='country')
clean_population_df
```
