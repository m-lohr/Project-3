```python
import pandas as pd
import json
import requests
```

```python
# load api
api_url = "https://restcountries.com/v3.1/all"
response = requests.get(api_url)
data = response.text
#response.json()
```

```python
# create dataframe
rest_countries = json.loads(data)
rest_countries_df = pd.json_normalize(rest_countries)

rest_countries_df.head()
```

```python
languages_df = rest_countries_df.filter(regex=('languages.*'))
languages_df
```

```python
population_df = rest_countries_df[['name.common','population']]
population_df.head()
```

```python
print(rest_countries_df['region'].unique())
print(rest_countries_df['subregion'].unique())
```

```python
rest_countries_df['continents'] = rest_countries_df['continents'].astype(str)
print(rest_countries_df['continents'].unique())
```

```python
world_map_df = rest_countries_df[['name.common','capital','region','subregion','continents']].copy()
world_map_df['continents'] = world_map_df['continents'].astype(str)
world_map_df['continents'] = world_map_df['continents'].str.replace('[','')
world_map_df['continents'] = world_map_df['continents'].str.replace(']','')
world_map_df['continents'] = world_map_df['continents'].str.replace("'",'')
world_map_df['capital'] = world_map_df['capital'].astype(str)
world_map_df['capital'] = world_map_df['capital'].str.replace('[','')
world_map_df['capital'] = world_map_df['capital'].str.replace(']','')
world_map_df['capital'] = world_map_df['capital'].str.replace("'",'')


world_map_df.head()
```

```python
world_map_df = world_map_df.assign(id=(world_map_df['region']).astype('category').cat.codes)
world_map_df = world_map_df.rename(columns={'id':'region_id'})
world_map_df = world_map_df.assign(id=(world_map_df['subregion']).astype('category').cat.codes)
world_map_df = world_map_df.rename(columns={'id':'subregion_id'})
world_map_df = world_map_df.assign(id=(world_map_df['continents']).astype('category').cat.codes)
world_map_df = world_map_df.rename(columns={'id':'continents_id', 'name.common':'country'})

world_map_df
```

```python
region_df = world_map_df[['region_id','region']]
region_df = region_df.drop_duplicates()
region_df
```

```python
subregion_df = world_map_df[['subregion_id','subregion']]
subregion_df = subregion_df.drop_duplicates()
subregion_df
```

```python
continents_df = world_map_df[['continents_id','continents']]
continents_df = continents_df.drop_duplicates()
continents_df
```

```python
world_map_df = world_map_df[['country','capital','region_id','subregion_id','continents_id']]
world_map_df
```

```python
# Convert dataframes to csv
world_map_df.to_csv('world_map.csv')
region_df.to_csv('region.csv')
subregion_df.to_csv('subregion.csv')
continents_df.to_csv('continents.csv')
```
